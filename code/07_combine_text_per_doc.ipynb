```python
''' 
combine the sentences of each report 
merge with report and company info
ran for each year 2011-2020
'''
```

```python
import pickle
import pandas as pd
import numpy as np
```

```python
year=2011
```

```python
with open('../data/preproc/preproc_text_' + str(year) + '.pickle', 'rb') as read_file:
    preproc_text = pickle.load(read_file)
    
preproc_text.shape  
```

```python
preproc_text.head()
```

```python
doc_df = preproc_text.groupby(['ticker','linkToTxt'])['preproc_text'].apply(' '.join).reset_index()
```

```python
doc_df.head()
```

```python
doc_df.tail()
```

```python
sp500_list = pd.read_pickle('../data/urls_' + str(year) + '.pickle')
sp500_list.head()
```

```python
#merge company info and disclosures
text_df_sp = pd.merge(doc_df, sp500_list, how="left", on=["ticker", "linkToTxt"])
text_df_sp.head()
```

```python
with open('../data/doc/doc_text_' + str(year) + '.pickle', 'wb') as to_write:
    pickle.dump(text_df_sp, to_write)
```
