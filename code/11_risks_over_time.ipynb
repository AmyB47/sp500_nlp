```python
"""
group data to plot relative frequency of risks over time
2011-2020
"""
```

```python
import pickle
import pandas as pd
import numpy as np
import datetime as dt
```

```python
with open('../data/topics/NMF_22_risk_topics_FINAL.pickle', 'rb') as read_file:
    risk_topics = pickle.load(read_file)
```

```python
risk_topics.head(2)
```

```python
# year column
risk_topics['year'] = risk_topics['date_filed'].dt.year
risk_topics.head(2)
```

```python
risk_topics.year.unique()
```

```python
risk_topics.columns
```

```python
year_agg =risk_topics[['year', 'risk_topic_nmf', 'risk_topics', 'value']].groupby(['year', 'risk_topic_nmf','risk_topics']).count().reset_index()
```

```python
year_agg.head()
```

```python
year_agg.tail()
```

```python
year_totals = year_agg[['year', 'value']].groupby(['year']).agg({'value':np.sum}).reset_index()
```

```python
year_totals.columns = ['year', 'year_total']
year_totals
```

```python
by_year = pd.merge(year_agg, year_totals, how='left', on='year')
by_year.head()
```

```python
by_year['freq'] = (by_year['value'] / by_year['year_total']) *100
by_year.head()
```

```python
by_year.head(25)
```

```python
by_year.to_excel('by_year_table.xlsx', index=False)
```

```python
pivot_year = by_year.pivot_table(index=['risk_topics'], columns='year', values='freq', fill_value=0).reset_index()
pivot_year.head()
```

```python
pivot_year.to_excel('pivot_year_table.xlsx', index=False)
```

```python
pivot_year2 = by_year.pivot_table(index=['year'], columns='risk_topics', values='freq', fill_value=0).reset_index()
pivot_year2.head()
```

```python
pivot_year2.to_excel('pivot_year_table2.xlsx', index=False)
```
